## Slice a tree
This IJulia notebook is to demonstrate how to slice an STL file. 

STL file is courtesy of Jordan Miller, and can be found [here](http://www.thingiverse.com/thing:31464).

```julia
using MeshSlicer
using PyPlot
```

```julia
mesh = PolygonMesh("jmil_2012_GeneratedVasculature.stl");
```

```julia
layers = 100;

# find slice heights
step = (mesh.bounds.zmax - mesh.bounds.zmin)/layers
slices = PolygonSlice[]

# create an array of slices
for i = mesh.bounds.zmin:step:mesh.bounds.zmax
    push!(slices, PolygonSlice(mesh,i))
end

# plot slices
for slice in slices
    for seg in slice.segments
        plot3D([seg.start[1],seg.finish[1]], [seg.start[2],seg.finish[2]], slice.layer)
    end
end
```
